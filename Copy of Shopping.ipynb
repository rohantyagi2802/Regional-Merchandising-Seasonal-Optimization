```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/customer_shopping_behavior.csv")
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
labels=['Young_Adult','Adult','Middle_Aged','Senior']
df['age_group']=pd.qcut(df['age'],labels=labels,q=4)
```

```python
df[['age','age_group']].head()
```

```python
frequency_mapping={'Fortnightly':14,'Weekly':7,'Monthly':30,'Quarterly':90,'Bi-Weekly':14,'Annually':365,'Every 3 months':90}
df['frequency_in_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_in_days','frequency_of_purchases']]
```

```python
(df['promo_code_used']==df['discount_applied']).all()
```

```python
df=df.drop(['promo_code_used'],axis=1)
```

```python
df.columns
```

```python
df['high_value_purchase_morethan70']=df['purchase_amount_(usd)']>70
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount_usd'})
```

```python
df.to_csv("clean_data.csv", index=False)
```
